- baby-weights-datasets.csv is the training set

- Objective: Predict the baby weight based off the given datasets.

```python
import pandas as pd
```

```python
training_set = pd.read_csv('datasets/baby-weights-dataset.csv')
```

```python
training_set.head()
```

```python
test_set = pd.read_csv('datasets/judge-without-labels.csv')
```

```python
test_set.head()
```

```python
training_set.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Apply LabelEncoder to the object columns
training_set['HISPMOM'] = le.fit_transform(training_set['HISPMOM'])
training_set['HISPDAD'] = le.fit_transform(training_set['HISPDAD'])

test_set['HISPMOM'] = le.fit_transform(test_set['HISPMOM'])
test_set['HISPDAD'] = le.fit_transform(test_set['HISPDAD'])
```

```python
training_set.dtypes
```

```python
# shuffle the training set
shuffled_training = training_set.sample(frac=1)
shuffled_training.head()
```

```python
# split into training (80%) and test (20%)
from sklearn.model_selection import train_test_split

X = shuffled_training.drop('BWEIGHT', axis=1)
y = shuffled_training['BWEIGHT']
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn import linear_model
import numpy as np
model = linear_model.LinearRegression()

# Convert numpy array to pandas DataFrame
# X_train_df = pd.DataFrame(X_train)

# # Drop rows with NaN values
# X_train_cleaned = X_train_df.dropna()
X_train_fixed = np.nan_to_num(X_train, nan=0)
# Fit the model with cleaned data
model.fit(X=X_train_fixed, y=y_train )


```

```python
prediction = model.predict(test_set)
```

```python
# Create a DataFrame with the predictions
predictions_df = pd.DataFrame(prediction, columns=['Predicted_BWEIGHT'])

# Save to CSV
predictions_df.to_csv('predicted_baby_weights.csv', index=False)
```

```python
def rmsle(predictions, targets):
    return np.sqrt(np.mean((np.log(predictions + 1) - np.log(targets + 1)) ** 2))
```

```python
predicted_weights = pd.read_csv('predicted_baby_weights.csv')
predictions = predicted_weights['Predicted_BWEIGHT']
predictions.head()
```

```python
target = training_set['BWEIGHT']
target.head()
```

```python
rmsle_error = rmsle(predictions, target)
print(f'RMSLE: {rmsle_error:.4f}')
```
